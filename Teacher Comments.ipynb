```python
import csv

with open('students.csv') as f:
    students = {}
    header = f.readline().split(',')
    data = csv.reader(f, delimiter=',')
    for row in data:
        students[row[0]] = (row[1:])


sep = "——————————————————————————————————————————"

# 'Student name ', 'Grade level 0', 'Pronouns 1',
# 'First Semester Grade 2', 'Second Semester Grade 3',
# 'Unit 1 Final 4', 'Unit 2 Final 5', 'Unit 3 Final 6',
# 'Participation Grade 7', 'Homework grade 8',
# 'Number of late or absents 9', 'Missing work 10',
# 'Special project 11', 'Group project 12',
# 'General comment 13', 'Special comment 14'


description = "Trigonometry is a hard subject for many students to learn, \
as its a bedrock subject for many future math classes, so any weakensses \
will come back in the future. Trig focuses on triangles, and the \
relationship between side lenghts of triangles, and the angles of \
triangles. Our class this year was especially difficult, as trig is a \
visual subject, and our class was entirely online."


def lg(a):
    '''returns a letter grade dependent on percent'''
    n = int(a)
    if n >= 100:
        return('A+')
    if 100 > n >= 93:
        return('A')
    if 93 > n >= 90:
        return('A-')
    if 90 > n >= 87:
        return('B+')
    if 87 > n >= 83:
        return('B')
    if 83 > n >= 80:
        return('B-')
    if 80 > n >= 77:
        return('C+')
    if 77 > n >= 73:
        return('C')
    if 73 > n >= 70:
        return('C-')
    if 70 > n >= 67:
        return('D+')
    if 67 > n >= 63:
        return('D')
    if 63 > n >= 60:
        return('D-')
    if 60 > n:
        return('F')


def finals(name):
    '''returns comment based on performance in final exams'''
    final_1 = int(students[name][4])
    final_2 = int(students[name][5])
    final_3 = int(students[name][6])
    one_two_dif = final_1 - final_2
    two_three_dif = final_2 - final_3
    if final_1 > 80:
        comment = "You did pretty well on your unit one final \
so good job on that,"
    elif final_1 > 60:
        comment = """Your unit one final could've used a lot of work,"""
    else:
        comment = """Your unit one final really brought your grade down,"""
    if one_two_dif == 0:
        comment += """ and your unit two final was the same."""
    elif one_two_dif < 0 and one_two_dif > -10:
        comment += """ and your unit two final was worse."""
    elif one_two_dif < -9:
        comment += """ and your unit two final was a lot worse."""
    elif one_two_dif > 0 and one_two_dif < 10:
        comment += """ and your unit two \
final was an improvement."""
    else:
        comment += """ and you significantly improved your score on your \
unit two final."""
    if two_three_dif < -10:
        comment += """ Your unit three final score was a lot worse from your \
unit two final."""
    elif two_three_dif < 0:
        comment += """ Your unit three final score was a bit lower than your \
unit two final."""
    elif two_three_dif < 10:
        comment += """ Your unit three final score was a bit better than your \
unit two final."""
    elif two_three_dif < 30:
        comment += """ Your unit three final score was significantly better \
than your unit two final."""
    else:
        comment += """ Your unit three final score was a complete turnaround \
from your unit two final."""
    return comment


def absents(name):
    '''returns comment based on number of absents'''
    absent_num = int(students[name][9].split("""/""")[1])
    if absent_num == 0:
        return """You never missed any classes without sending in a note \
or emailing to excuse yourself first."""
    elif absent_num < 3:
        return """You missed class once or twice but it's not too big of \
a deal, just don't do it again."""
    elif absent_num < 7:
        return """You missed class a couple times so I'm not sure what was \
up with that, and I did have to mark your grade down a bit because \
of it."""
    elif absent_num > 6:
        return """You frequently missed class, and it severely affected your \
grade."""


def lates(name):
    '''returns comment based on number of tardies'''
    late_num = int(students[name][9].split("""/""")[0])
    if late_num == 0:
        return """You were also on time every class and even occasionally \
got there early, so I appreciate that."""
    elif late_num < 3:
        return """You also were late a couple times but that's ok."""
    elif late_num < 7:
        return """You were also late a bunch of times, so I had to mark you \
down for that. It's definitely an area for improvement."""
    elif late_num > 6:
        return """You were also frequently late to class which I'll definitely \
need you to change going forward."""


def growth(name):
    '''returns a comment dependent on growth from semester one \
to semester 2'''
    grade_comparison = int(students[name][3]) - int(students[name][2])
    """takes in the students name and returns whether they improved in \
the second semester from their first semester grade"""
    if grade_comparison == 0:
        return """Your first and second semester were the same."""
    elif grade_comparison < 0:
        return """Your grade overall worsened in the second semester."""
    elif grade_comparison > 0:
        return """Your grade overall improved in the second semester."""


def participation(name):
    '''this function takes in the students name and returns a comment based \
off their participation grade'''
    pg = int(students[name][7])
    if pg >= 95:
        return("""You participated actively in this class, contributing \
often to class conversations and working well with other students.""")
    elif 90 <= pg < 95:
        return("""You particpated often in class, but a bit more preparation \
could bring these conversations to the next level.""")
    elif 85 <= pg < 90:
        return("""You offered some good comments in class, but a you could've \
included more detail and kept on topic.""")
    elif 80 <= pg < 85:
        return("""You didn't talk as much as you probably should have in \
class, and when you did you didn't seem very engaged.""")
    elif 70 <= pg < 80:
        return("""You didn't pay attention during class, and seemed more \
interested in your phone and computer than the discussions.""")
    elif pg < 70:
        return("""You didn't even try to pretened to pay attention, and \
would talk over me during class.""")


def sem1(name):
    '''this function takes in the students name and returns a comment \
based off their behavior in the first semester'''
    s1 = int(students[name][2])
    if s1 >= 95:
        return("""First semester you showed an in depth, precise \
understanding of trig, you really achieved that next level of \
understanding for the first couple units.""")
    elif 90 <= s1 < 95:
        return("""For the first semester you showed a good understanding \
of the math we were learning, and while you struggled with a \
couple areas you managed to overcome these issues.""")
    elif 85 <= s1 < 90:
        return("""Your work in the first semester was well done, although \
you struggled with some stuff, and you might've met with me to go \
over them, you understand the big picture topics.""")
    elif 80 <= s1 < 85:
        return("""You struggled first semester to accurately show your \
understanding of the math we learned, and I was hoping you might \
have met with me one on one to go over some of the more difficult \
topics.""")
    elif 70 <= s1 < 80:
        return("""First semester was very difficult for you, and while you \
understood some of the units we learned, you didn't put in as much \
time or effort as you could've.""")
    elif s1 < 70:
        return("""You failed first semester, while I understand trig is \
hard you didn't demonstrate any knowledge of trig.""")


def sem2(name):
    '''this function takes in the students name and returns a comment \
based off their behavior in the second semester'''
    s2 = int(students[name][3])
    if s2 >= 95:
        return("""You really excelled this year in trig, demonstrating \
clear understanding of the units we covered, with nice work \
shown and care for details.""")
    elif 90 <= s2 < 95:
        return("""And while you did very well second semester, you \
could've payed more attention to some of the details, this \
being said you clearly understand trig.""")
    elif 85 <= s2 < 90:
        return("""And you understand the majority of the content \
we've covered this year. However, that understanding \
didn't always translate into your work.""")
    elif 80 <= s2 < 85:
        return("""This being said, your work second semester was \
for the most part complete, but you didn't understand \
enough of the concepts to finish everything, and you \
didn't push your grade to the next level.""")
    elif 70 <= s2 < 80:
        return("""With this in mind, I was hoping you would meet with me \
a couple times second semester to help improve your understanding \
of these concepts, but you didn't, and your grade suffered for it.""")
    elif s2 < 70:
        return("""Keeping this in mind, you didnt put in much work to my \
class at all, and you demonstrated little understanding of the \
concepts we covered. I was dissapointed when I had to fail you.""")


def group_work(name):
    '''this function takes in the students name and returns a comment \
based off their group work grade'''
    gw = int(students[name][12])
    if gw >= 95:
        return(f"""You clearly work great in small groups, as you got a \
{students[name][12]} on the group project. You put in a ton of \
time making sure everyone got their work done, and you helped \
keep the conversations focused on classwork.""")
    elif 90 <= gw < 95:
        return(f"""You did a nice job working within smaller groups, \
and you displayed this during the group project where you got \
a {students[name][12]}.""")
    elif 85 <= gw < 90:
        return(f"""When working with smaller groups in class you clearly \
helped them out, but I wish you might have put in bit more work \
into the group project where you got a {students[name][12]}.""")
    elif 80 <= gw < 85:
        return(f"""While you got the work done when working with others in \
a small group, some students said that you might relied a bit too \
heavlily on the work of others, and didn't contribute enough yourself. \
This occured during the group project where you got a \
{students[name][12]}.""")
    elif 70 <= gw < 80:
        return(f"""During work time within small groups you didn't seem \
too interested in helping out those around you, and instead seemed \
to enjoy talking to them about activies outside of school. I think \
this is most evident during our group project, where you got a \
{students[name][12]}. This is definitely a major area for improvement \
for you.""")
    elif gw < 70:
        return(f"""I was dissapointed by your lack of respect for your peers \
when it came to small group work, as you clearly assumed they would \
do the majority of the work. You got a {students[name][12]} on the \
group project, which definitely did not help your grade.""")


def missing_work(name):
    '''this function takes in the students name and returns a comment based
    off the amount of their missing work'''
    mw = int(students[name][10])
    if mw >= 10:
        return("You had a lot of missing work this year, ")
    if 10 > mw >= 5:
        return("You had a decent amount of missing work this year, ")
    if 5 > mw >= 0:
        return("You had little or no missing work this year, ")


def homework(name):
    '''this function takes in the students name and returns a comment based
    off their homework grade'''
    hw = int(students[name][8])
    if hw >= 95:
        return("""and the work you turned in was clean, with work \
shown and nice notes.""")
    if 95 > hw >= 90:
        return("""and the work you turned in was well done, although you \
could've focused on the details a bit more.""")
    if 90 > hw >= 85:
        return("""and the work you turned in for the most part was well \
done, although you didn't always make the most out of homework.""")
    if 85 > hw >= 80:
        return("""and the work you turned in while present for the most \
part, didn't help you strengthen those skills we wentover in class.""")
    if 80 > hw >= 70:
        return("""and the work you did turn in was pretty sloppy and \
felt incomplete.""")
    if 70 > hw >= 60:
        return("""which impacted your grade, and the work you turned \
in felt rushed and lacked work shown.""")
    if 60 > hw:
        return("""which heavily impacted your grade, on top of the \
little work you showed on the homework you decided to turn in.""")


def best_aspect(name):
    grade_list = []
    counter = 4
    for i in students[name][4:8]:
        grade_list.append((int(i), header[counter]))
        counter += 1
    grade_list.append((int(students[name][11]), header[11]))
    grade_list.append((int(students[name][12]), header[12]))
    grade_list.sort(reverse=True)
    if grade_list[0][0] > 90:
        word = "so good job on that."
    else:
        word = "but that wasn't too high so you still have a lot to work on."
    return f"Your highest score this year was \
{grade_list[0][0]} on your {grade_list[0][1]}, {word}"


def grade_level(name):
    gl = int(students[name][0])
    if gl == 12:
        return(f"""I have to run to another conference so good luck \
in college {name.split(' ')[0]}, and I'm sure you'll do great things.""")
    else:
        return(f"""Well that's all the time that I have so I hope to \
see you next year {name.split(' ')[0]}, and I'm \
excited to watch you grow as a learner.""")


def final_comment(name):
    file = open(f"{name}.txt", 'w')
    file.write(f"""Student Name: {name}\nGrade level: {students[name][0]} \
\nSemester 1 Grade: {lg(students[name][2])} \
({students[name][2]}%)\nSemester 2 Grade: {lg(students[name][3])} \
({students[name][3]}%)\n\n{description}\n\n{sem1(name)} \
{growth(name)} {sem2(name)} {finals(name)}\n
{participation(name)} {group_work(name)}\n\n{missing_work(name)} \
{homework(name)}\n\n{absents(name)} {lates(name)}\n
{students[name][13]} {best_aspect(name)}\n\n{grade_level(name)}""")
    file.close()
```

```python
mode = input("Do you want one student, or all? students?\
(one/all)")
if mode == "all":
    for i in students:
        final_comment(i)
    shown = input("Do you want this comment displaye\
d? (yes/no)")
    if shown == 'yes':
        for i in students:
            with open(f"{i}.txt") as f:
                for line in f:
                    print(line.strip())
                print(f"\n{sep}\n")
elif mode == "one":
    single = input("Which one? (use their name)")
    if single in students:
        final_comment(single)
        shown = input("Do you want this comment displaye\
d? (yes/no)")
        if shown == 'yes':
            with open(f"{single}.txt") as f:
                for line in f:
                    print(line.strip())
    else:
        print("I don't recognize that student, I'm sorry.")
```
